```python
import scipy.stats as stats

print(df.shape)
print(list(df.columns))
print(df.head())

#Clean data
# Handle missingness
nan_count = np.sum(df.isnull(), axis = 0)
print(nan_count)
df.drop(columns = 'GINI index (World Bank estimate)', inplace=True) #too much Null fields
df.drop(columns = 'country', inplace=True) #irrelevant and categorical value 
nan_count = df.isnull().sum()
print(df.dtypes)

# Replace null field with mean value (for float/int type)
to_impute_selected = df.columns[(nan_count != 0) & ((df.dtypes == 'int64') | (df.dtypes == 'float64'))]
for colname in to_impute_selected:
    df[colname +'_na'] = df[colname].isnull()
for colname in to_impute_selected:
    df[colname].fillna(np.mean(df[colname]), inplace=True)
for colname in to_impute_selected:
    print("{} missing values count :{}".format(colname, np.sum(df[colname].isnull(), axis = 0)))
print(df.columns)
df = df.drop(columns=df.select_dtypes(include='bool').columns)

# Handle outliers
df['label_life_ladder'] = stats.mstats.winsorize(df['Life Ladder'], limits=[0.01, 0.01])
print("Original vs Winsorized differences in label:", (df['Life Ladder'] - df['label_life_ladder']).unique())

# Handle irrelevant features
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation with the label
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
target_corr = corr_matrix['Life Ladder'].drop('Life Ladder')

target_corr_sorted = target_corr.reindex(target_corr.abs().sort_values(ascending=True).index)
print("Correlation with 'Life Ladder':\n", target_corr_sorted)

to_drop = target_corr_sorted.index[:3]
df = df.drop(columns=to_drop)

plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=True)
plt.title("Correlation Matrix")
plt.show()

df.columns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

<b>Task:</b> Use the rest of this notebook to carry out your project plan. 

You will:

1. Prepare your data for your model.
2. Fit your model to the training data and evaluate your model.
3. Improve your model's performance by performing model selection and/or feature selection techniques to find best model for your problem.

Add code cells below and populate the notebook with commentary, code, analyses, results, and figures as you see fit. 

```python
# Define label
y = df['label_life_ladder']
X = df.drop(columns=['Life Ladder', 'label_life_ladder'], axis=1)

#Train and split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1234)

# Train, fit, and evaluate model
rf_model = RandomForestRegressor(n_estimators=200, max_depth=20, min_samples_split=2)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# Feature importance plot 

importances = rf_model.feature_importances_
features = X.columns

plt.figure(figsize=(10, 6))
pd.Series(importances, index=features).sort_values().plot(kind='barh')
plt.title("Feature Importances")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()

# Tunning hyperparameters for model selections
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [1, 10, 20],
    'min_samples_split': [2, 5],
}

grid_search = GridSearchCV(RandomForestRegressor(random_state=1234), param_grid, cv=3, scoring='r2')
grid_search.fit(X_train, y_train)

print("Best parameters from GridSearchCV:", grid_search.best_params_)

# Evaluate best model from tuning
best_model = grid_search.best_estimator_
y_pred_best = best_model.predict(X_test)
print(f"Tuned R^2 Score: {r2_score(y_test, y_pred_best):.2f}")
```
